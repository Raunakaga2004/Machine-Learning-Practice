### Handling Missing Data

#### for cabin column

```python
df['Cabin'].isnull().sum()/df.shape[0] *100

# therefore more than 70% data is missing in this column
# So if more than 50% data is missing in a column then it is recommended to remove it
```

```python
cabin = df['Cabin'].copy() # for further use in handling mixed variables
```

```python
df.drop('Cabin', axis = 1, inplace = True)
```

```python
df.head()
```

#### for age column

```python
df['Age'].isnull().sum() / df.shape[0] *100

# around 20% data is missing in this column, we can not apply CCA (removing rows)
# around 20% we can consider imputing the column
```

```python
# as Age is a numerical column we can apply mean and median imputation
df_Age1 = df['Age'].fillna(df['Age'].mean())
df_Age2 = df['Age'].fillna(df['Age'].median())
```

```python
import seaborn as sns

sns.kdeplot(data = df, x = 'Age', color="blue")
sns.kdeplot(x = df_Age1, color= "red")
sns.kdeplot(x = df_Age2, color ="green")

# they have high kurtosis
```

```python
print(df['Age'].std())
print(df_Age1.std())
print(df_Age2.std())

# std deviation is almost same
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(12, 4))

sns.boxplot(data = df, x = 'Age', color="blue", ax = axes[0])
sns.boxplot(x = df_Age1, color= "red", ax = axes[1])
sns.boxplot(x = df_Age2, color ="green", ax = axes[2])

# number of outliers are increased very much
```

```python
import numpy as np
import random

def  random_Imputation(x) :
    if np.isnan(x) :
      return random.choice(list(df.Age.dropna()))
    else :
      return x
df_Age3  = df['Age'].apply(random_Imputation)
```

```python
import seaborn as sns

sns.kdeplot(data = df, x = 'Age', color="blue")
sns.kdeplot(x = df_Age3, color= "red")

# so in this the distribution is not changed
```

```python
print("Original age correlation : ", df['Pclass'].corr(df['Age']))
print("Random imputed age correlation : ", df['Pclass'].corr(df_Age3))

# correlation between other column is increased little bit
```

```python
print(df['Age'].std())
print(df_Age3.std())

# std deviation is same
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

sns.boxplot(data = df, x = 'Age', color="blue", ax = axes[0])
sns.boxplot(x = df_Age3, color= "red", ax = axes[1])

# number of outliers are also almost same as the distribution is same
```

```python
# using multivariate imputation (KNN imputer)

from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=5, weights = "distance")
df_Age4 = imputer.fit_transform(df[['Age']])

df_Age4 = pd.DataFrame(df_Age4, columns = ["Age"])
df_Age4
```

```python
import seaborn as sns

sns.kdeplot(data = df, x = 'Age', color="blue")
sns.kdeplot(x = df_Age4['Age'], color= "red")
```
